```python
import random 
def walk1D(x0,N,d):
    X=[]
    T=[]
    for t in range(N):
        x= x0+random.choice([-1,1])*d
        T.append(t)
        X.append(x)
    return T,X
```

```python
walk1D(0,10,1)
```

```python
import matplotlib.pyplot as plt
time,pos=walk1D(0,100,1)
```

```python
plt.plot(time,pos)
```
